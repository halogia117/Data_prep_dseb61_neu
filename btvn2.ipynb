### Ex1 : Calculate the mean of mag in Alaska and the distance is smaller than 100km, example: "62km W of Cantwell, Alaska".


```python
df1 = df[['mag', 'title']]
df11 = df1['title'].str.split(" ", n = 8, expand = True)
df1[['km', 'locc']] = df11[[3, 7]];
df12 = df1['km'].str.split("k", n = 2, expand = True)
df1[['km_num']] = df12[[0]]
```

```python
#df1 = df1.drop('title', axis = 1)
df1['km_num'] = pd.to_numeric(df1['km_num'], errors='coerce')
df1.info()
```

```python
final_df = df1[(df1['locc'] == 'Alaska') & (df1['km_num'] < 100)]
final_df
```

```python
np.mean(final_df['km_num'])
```

### Ex3 : Calculate the mean of mag in each quarter (q1:m1-m3, q2: m4-m6, q3: m7-m9, q4: m10-12), the "time" field stores the milliseconds from 1/1/1970, hint: timestamp.

```python
df111 = df[['mag', 'time']]
df111['time_num'] = pd.to_numeric(df111['time'], errors='coerce')
df111["time_num"] = pd.to_datetime(df111["time_num"], unit='ms').dt.date
```

```python
df111["month"] = df111['time_num'].map(lambda x: x.month)
df111["year"] = df111['time_num'].map(lambda x: x.year)
df_fin = df111.drop(['time', 'time_num'], axis = 1)
```

```python
def check_q(x):
    a =  0
    if 1 <= x <= 3:
        a = 1
    elif 4 <= x <= 6:
        a = 2
    elif 7 <= x <= 9:
        a = 3
    else:
        a = 4
    return a
```

```python
df_fin["quater"] = df_fin['month'].map(lambda x: check_q(x))
df_fin
```

```python
df_fin.info()
```

```python
for i in range(4):
    print('mean_mag quater', i+1 ,' is :', np.mean(df_fin[df_fin['quater'] == (i+1)].mag))
```
