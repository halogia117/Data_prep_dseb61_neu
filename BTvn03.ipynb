## Ex1 : We want to look at data for the Facebook, Apple, Amazon, Netflix, and Google (FAANG) stocks. Combine them  into a single file and store the dataframe of the FAANG data as faang for the rest of the exercises:

### A, Read in the aapl.csv, amzn.csv, fb.csv, goog.csv, and nflx.csv files.

```python
AAPL = pd.read_csv('aapl.csv')
AMZN = pd.read_csv('amzn.csv')
FB = pd.read_csv('fb.csv')
GOOG = pd.read_csv('goog.csv')
NFLX = pd.read_csv('nflx.csv')
```

### B, Add a column to each dataframe, called ticker, indicating the ticker symbol it is for (Apple's is AAPL, for example); this is how you look up a stock. In this case, the filenames happen to be the ticker symbols. 

```python
AAPL['Ticker'] = 'AAPL'
AMZN['Ticker'] = 'AMZN'
FB['Ticker'] = 'FB'
GOOG['Ticker'] = 'GOOG'
NFLX['Ticker'] = 'NFLX'
```

### C, Append them together into a single dataframe. (concat)

```python
df1 = pd.concat([AAPL, AMZN, FB, GOOG, NFLX])
```

```python
df1
```

### D, save the result in a CSV file called faang.csv. (to_csv)

```python
df1.to_csv('faang.csv')
```

### E, With faang, use type conversion to cast the values of the date column into datetimes and the volume column into integers. Then, sort by date and ticker.

```python
faang = df1 
```

```python
faang.info()
```

```python
faang['date'] = pd.to_datetime(faang['date'])
faang
```

```python
faang['volume'] = pd.to_numeric(faang['volume'])
```

```python
faang.info()
```

```python
faang.sort_values(by=['date', 'Ticker'])
```

### F, Find the seven rows in faang with the lowest value for volume

```python
faang.sort_values(by=['volume']).head(7)
```

## Ex2: The European Centre for Disease Prevention and Control (ECDC) provides an open dataset on COVID-19 cases called  daily number of new reported cases of COVID-19 by country worldwide. This dataset is updated daily, but we will use a snapshot that contains data from January 1, 2020 through September 18, 2020. Clean and pivot the data so that it is in wide format: 

### A, Read in the covid19_cases.csv file.

```python
covid = pd.read_csv('covid19_cases.csv')
```

```python
covid.tail(3)
```

### B, Create a date column using the data in the dateRep column and the pd.to_datetime() function.

```python
covid['date'] = pd.to_datetime(covid['dateRep'])
covid.tail(3)
```

### C, Set the date column as the index and sort the index.

```python
#covid = covid.set_index('date')
covid.sort_index()
```

### D, Replace all occurrences of United_States_of_America and United_ Kingdom with USA and UK, respectively. Hint: the replace() method can be run on the dataframe as a whole.

```python
covid['countriesAndTerritories'] = covid['countriesAndTerritories'].replace({'United_States_of_America':'USA'})
covid['countriesAndTerritories'] = covid['countriesAndTerritories'].replace({'United_Kingdom' :'UK'})
```

```python
covid['countriesAndTerritories'].unique()
```

### E, sing the countriesAndTerritories column, filter the cleaned COVID-19 cases data down to Argentina, Brazil, China, Colombia, India, Italy, Mexico, Peru, Russia, Spain, Turkey, the UK, and the USA. 

```python
Country = ['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey','UK', 'USA']
```

```python
df2 = covid[covid['countriesAndTerritories'].isin(Country)]
```

```python
df2['countriesAndTerritories'].unique()
```

```python
df2 = df2.set_index('date')
```

```python
df2.sort_index()
df2
```

### F, Pivot the data so that the index contains the dates, the columns contain the country names, and the values are the case counts (the cases column). Be sure to fill in NaN values with 0.

```python
df3 = covid[covid['countriesAndTerritories'].isin(Country)]
pivot_df = df3.pivot(
    index = 'date', columns = 'countriesAndTerritories', values = 'cases'
)
```

```python
pivot_df
```

```python
pivot_df.fillna(0)
```
