### Ex1 Load dataset nyc_weather_2018 and weather_stations, then calculate the average elevation of stations that calculat‘SNOW’ and contains ‘NJ US’ in its name.

```python
import pandas as pd
import numpy as np
```

```python
weather = pd.read_csv("./nyc_weather_2018.csv")
station_infor = pd.read_csv("./weather_stations.csv")

station_infor.tail(15)
```

```python
station_infor.tail(15)
```

```python
station_infor.rename(columns={"id":"station"},inplace=True)
station_infor.tail(10)
```

```python
data = station_infor.merge(weather,on="station",how='inner') \
                            .drop_duplicates(["station","datatype"]).reset_index(drop=True)
data.shape
```

snow_avg= data.query('datatype=="SNOW" and name.str.contains("NJ US")',engine="python")
snow_avg.elevation.mean()

### Ex2: Load the fb dataset, then:
 •Calculate the percentage change between that and and last day in volumn (pct_change method)
 
 •Sort the dataframe by percentage change then get five days that have the biggest change.

```python
fb_2018 = pd.read_csv("./fb_2018.csv")
```

```python
fb_p = fb_2018.assign(p_change = lambda x: x.volume.pct_change())
fb_p.tail(10)
```

```python
fb_2018['shift_vol']=fb_2018.volume.shift(1)
fb_2018.head()
```

```python
fb_p2 = fb_2018.assign(p_change = lambda x: (x.volume.sub(x.shift_vol)).div(x.shift_vol))
fb_p2.head()
```

```python
fb_p2.sort_values(by=("p_change"),ascending=False).head()
```

### Ex3 : Create bins for each full number of earthquake magnitude (for instance, the first bin is (0, 1], the second is (1, 2], and so on) with the ml magnitude type and count how many are in each bin.

```python
earthquakes = pd.read_csv("earthquakes.csv")
earthquakes.tail(4)
```

```python
mag_ml = earthquakes[earthquakes.magType=="ml"]
earthquakes = earthquakes.assign(mag_bin = pd.cut(mag_ml.mag,bins=5,labels=["nor","low","med","high","dangerous"]))
earthquakes.mag_bin.value_counts()
```

### Ex4 : Load the dataset weather_by_station then:
    •Calculate the mean value each station recorded snow per month.
    •Count the number of times each station recorded snow per month.

```python
weather_by_station = pd.read_csv("weather_by_station.csv")
```

```python
weather_by_station = weather_by_station.assign(month = pd.DatetimeIndex(weather_by_station.date).month)
```

```python
pd.crosstab(index = weather_by_station.station, columns = weather_by_station.month, colnames=['month'], values = weather_by_station.value, aggfunc=np.mean)
```
